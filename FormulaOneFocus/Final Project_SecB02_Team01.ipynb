```
#Reading the csv file into dataframe
df= spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/tables/final.csv")
```

```
#dispalying the dataframe
display(df)
```

```
#Printing the schema
df.printSchema()
```

```
#Converting the data into necessary data types
from pyspark.sql.types import StringType, FloatType
  
df = df \
  .withColumn("driverStandingsId" ,
              df["driverStandingsId"]
              .cast(FloatType()))   \
  .withColumn("driverId" ,
              df["driverId"]
              .cast(FloatType()))   \
  .withColumn("wins" ,
              df["wins"]
              .cast(FloatType()))   \
  .withColumn("points",
              df["points"]
              .cast(FloatType()))    \
  .withColumn("position"  ,
              df["position"]
              .cast(FloatType())) \
   .withColumn("duration"  ,
              df["duration"]
              .cast(FloatType())) \
      
df.printSchema()
```

```
#Filtering the necessary columns into the dataframe
df_1=df.select('name', 'wins','points','position','duration')
```

```
#Dispalying the necessary features using scatter matrix
from pandas.plotting import scatter_matrix

numeric_features = [t[0] for t in df_1.dtypes if t[1] == 'int' or t[1] == 'float']
sampled_data = df_1.select(numeric_features).sample(False, 0.8).toPandas()
axs = scatter_matrix(sampled_data, figsize=(10, 10))
n = len(sampled_data.columns)
for i in range(n):
    v = axs[i, 0]
    v.yaxis.label.set_rotation(0)
    v.yaxis.label.set_ha('right')
    v.set_yticks(())
    h = axs[n-1, i]
    h.xaxis.label.set_rotation(90)
    h.set_xticks(())
```

```
#From the above graph it is evident that there is no linear relationship between the variables.
```

```
#Displaying correlation coefficient of matrices
import six
for i in df_1.columns:
    if not( isinstance(df_1.select(i).take(1)[0][0], six.string_types)):
        print( "Correlation to wins for ", i, df_1.stat.corr('wins',i))
```

```
#Merging the feature columns into a vector
from pyspark.ml.feature import VectorAssembler
vectorAssembler = VectorAssembler(inputCols = ['points','position','duration'], outputCol = 'features')
t_df = vectorAssembler.transform(df_1)
t_df = t_df.select(['features', 'wins'])
t_df.show(3)
```

```
#Splitting the data into train and test sets
splits = t_df.randomSplit([0.7, 0.3])
train_df = splits[0]
test_df = splits[1]
```

```
#Training the test data and displaying the coefficients required for the models
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features', labelCol='wins', maxIter=10, regParam=0.3, elasticNetParam=0.8)
lr_model = lr.fit(train_df)
print("Coefficients: " + str(lr_model.coefficients))
print("Intercept: " + str(lr_model.intercept))
```

```
#Displaying the summary of the linear model
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```
#Displaying the predictions of the model using test set.
lr_predictions = lr_model.transform(test_df)
lr_predictions.select("prediction","wins","features").show(5)
from pyspark.ml.evaluation import RegressionEvaluator
lr_evaluator = RegressionEvaluator(predictionCol="prediction", \
                 labelCol="wins",metricName="r2")
print("R Squared (R2) on test data = %g" % lr_evaluator.evaluate(lr_predictions))
```

```
#Displaying the test result
test_result = lr_model.evaluate(test_df)
print("Root Mean Squared Error (RMSE) on test data = %g" % test_result.rootMeanSquaredError)
```

```
#Displaying the iterations and objective history
print("numIterations: %d" % trainingSummary.totalIterations)
print("objectiveHistory: %s" % str(trainingSummary.objectiveHistory))
trainingSummary.residuals.show()
```

```
#Displaying the predictions using lr model
predictions = lr_model.transform(test_df)
predictions.select("prediction","wins","features").show()
```

```
#Using the decision tree regressor
from pyspark.ml.regression import DecisionTreeRegressor
dt = DecisionTreeRegressor(featuresCol ='features', labelCol = 'wins')
dt_model = dt.fit(train_df)
dt_predictions = dt_model.transform(test_df)
dt_evaluator = RegressionEvaluator(
    labelCol="wins", predictionCol="prediction", metricName="rmse")
rmse = dt_evaluator.evaluate(dt_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```

```
#Displaying the feature importances
dt_model.featureImportances
```

```
#Using Gradient boost regressor
from pyspark.ml.regression import GBTRegressor
gbt = GBTRegressor(featuresCol = 'features', labelCol = 'wins', maxIter=10)
gbt_model = gbt.fit(train_df)
gbt_predictions = gbt_model.transform(test_df)
gbt_predictions.select('prediction', 'wins', 'features').show(5)
```

```
#Using GBT evaluator to evaluate the model
gbt_evaluator = RegressionEvaluator(
    labelCol="wins", predictionCol="prediction", metricName="rmse")
rmse = gbt_evaluator.evaluate(gbt_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```
